```python
import numpy as np
import math 
from random import randint
```

```python
class Genetic_Algorithm:

	populasi_individu = 6 
	individu = ["a", "b", "c", "d"] 
	batasan = {
		'min' : 0,'max' : 10
	} 
	CO_Rate = 0.25 
	M_Rate = 0.1 
	total_generasi = 100
	next_gen = np.arange(4)
	stop = False
	def __init__(self):
		self.chromosome = np.random.randint(self.batasan['min'], self.batasan['max'], size=(self.populasi_individu, len(self.individu)))
		print(self.chromosome)
		print("------------------------------------------------")

		

	def proses(self, chrom, generasi):
		print("Generasi ke ["+str(generasi)+"]")
		
		populasi_individu = len(chrom)
		j = np.arange(populasi_individu)   
		fitness = np.arange(populasi_individu, dtype='f')
		
		for x in range(len(chrom)):
			
			fo = abs((chrom[x][0]+2*chrom[x][1]+3*chrom[x][2]+4*chrom[x][3])-30)
			j[x] = fo
			

			fitn = 1/(fo+1)
      		
			fitness[x] = fitn
			if(fitn == 1):
				self.stop = True
			print("Chromosome {0} : {1}, fitness = {2}".format(x, np.array2string(chrom[x], separator=','), fitn))
		print("Fitness Selesai")
		print(j)
		
		P = np.arange(populasi_individu, dtype='f')
		P = fitness / fitness.sum()
		print("Rata-rata Fungsi Objektif : {}".format(str(np.average(fo))))
		print("Total Fitness : {}".format(str(fitness.sum())))
		print("Probabilitas : {}".format(np.array2string(P, separator=',')))
		print("Probabilitas tertinggi : {}, pada chromosome {}".format(P[P.argmax()], str(P.argmax())))
		print("Chromosome Terpilih Sementara  : {}".format(np.array2string(chrom[P.argmax()], separator=',')),end="\n\n") 

		self.next_gen = chrom[P.argmax()]
		C = np.arange(populasi_individu, dtype='f')
		total_x = 0
		for x in range(len(P)):
			total_x += P[x]
			C[x] = total_x

		R = np.random.sample(len(fitness))
		new_chrom = np.arange(populasi_individu*len(self.individu)).reshape(populasi_individu, len(self.individu))
		
		for y in range(len(R)):
			for k in range(len(new_chrom)):
				if(R[y] < C[0]):
					new_chrom[y] = chrom[0]
				elif((C[k-1] < R[y]) & (R[y] < C[k])):
					new_chrom[y] = chrom[k]
					
		
		R = np.random.sample(populasi_individu)
		index_chrom_parent = [] 
		for p in range(len(R)):
			if(R[p] < self.CO_Rate):
				index_chrom_parent.append(p)

		
		posisi_CO = np.random.randint(low=1, high=len(self.individu), size=len(index_chrom_parent))

		 
		off_spring = np.arange(len(self.individu)*len(index_chrom_parent)).reshape(len(index_chrom_parent), len(self.individu))

		for i_parent in range(len(index_chrom_parent)):
			index_chrome_1 = index_chrom_parent[i_parent]
			if(i_parent == len(index_chrom_parent)-1):
				index_chrome_2 = index_chrom_parent[0]
			else:
				index_chrome_2 = index_chrom_parent[i_parent+1]
			
			cut_point = posisi_CO[i_parent]
			for p in range(len(new_chrom[index_chrome_1])):
				
				if(p >= posisi_CO[i_parent]):
					off_spring[i_parent][p] = new_chrom[index_chrome_2][p]
				else:
					off_spring[i_parent][p] = new_chrom[index_chrome_1][p]

		for x in range(len(off_spring)):
			new_chrom[index_chrom_parent[x]] = off_spring[x]


		
		total_individu = len(chrom) * len(chrom[0])
		jumlah_mutasi = self.M_Rate * total_individu
		jumlah_mutasi = int(jumlah_mutasi)

		random_i_mutasi = np.random.randint(low=0, high=total_individu, size=jumlah_mutasi)

		for x in range(len(random_i_mutasi)):
			index_mutasi = random_i_mutasi[x]
			banyak_kromosom = len(chrom)
			banyak_individu = len(chrom[0]) 
			if(index_mutasi <= banyak_individu):
				new_chrom[0][index_mutasi-1]
			else:
				pos_y = index_mutasi/banyak_individu
				pos_y = int(pos_y)
				pos_x = index_mutasi % banyak_individu
				new_chrom[pos_y][pos_x] = np.random.randint(self.batasan['min'], self.batasan['max'])
		return new_chrom

	def do_now(self):
		chromosome_current = self.chromosome
		for generasi in range(0, self.total_generasi):
			if(self.stop != True):
				chromosome_current = self.proses(chromosome_current, generasi)
		print("---Selesai---")
		print("Chromosome Tertinggi Adalah")
		print(self.next_gen)

run = Genetic_Algorithm()
run.do_now()
```
